### Завдання 2

Програма приймає на вхід два параметри: стать спортсмена(`sex = (f, m)`) та вік спортсмена(`age = [0,100]`).  Спортсмен-дівчина може брати участь у змаганні, якщо її вік більший 18, спортсмен-хлопець - якщо вік більший 16

Якщо введеніть не коректні дані, виведіть повідомлення про помилку

```python
try:
    sex = input()
    age = float(input())

    if (sex == 'f') or (sex == 'm'):
        if (age > 0) and (age <= 100):
            if (sex == 'f') & (age >= 18) or (sex == 'm') & (age >= 16):
                print(True)
            else:
                print(False)
        else:
            print('Неправильно введені дані')
    else:
        print('Неправильно введені дані')
    
except Exception:
    print('Неправильно введені дані')
```

### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random

#генеруэмо сто чисел в діапазоні від 0 до 100
list_with_random_numbers = [random.randint(0,100) for i in range(100)]
print(list_with_random_numbers)
```

```python
def max_min():
    m = max(list_with_random_numbers)
    n = min(list_with_random_numbers)
    print(m)
    print(n)

max_min()
```

```python
def avg():
    avg = sum(list_with_random_numbers) / len(list_with_random_numbers)
    print(avg)
    
avg()
```

```python
def med():
    med = (list_with_random_numbers[49] + list_with_random_numbers[50])/2
    print(med)
    
med()
```

```python
dict_with_nums = {}

for n in list_with_random_numbers:
     dict_with_nums[n] = dict_with_nums.get(n, 0) + 1

print(dict_with_nums)
```

```python
import math

def sigma():
    sigma = math.sqrt((sum((n - avg) ** 2)) // len(list_with_random_numbers))
    print(sigma)
    
sigma()
```
